```python
import pandas as pd
from experiment.validacao import CalculoMetricas
```

```python
modelos = [
    'ExtremelyFastDecisionTreeClassifier',
    'HoeffdingAdaptiveTreeClassifier',
    'HoeffdingTreeClassifier',
    'AMFClassifier',
    'ARFClassifier',
    # 'ALMAClassifier',
    'KNNClassifier'
]
```

```python
dados_tudo = []
for m in modelos:
    dados = []
    for i in range(4):
        df_REC = pd.read_csv(f'../experiment/modelo_treinado/{m}_cenario_{i}/recomendacao_incremental_model.csv')
        df_REC['X_index'] = pd.to_datetime(df_REC['X_index'])
        df_REC.set_index('X_index', inplace=True)

        df_X_Y = pd.read_csv(f'../experiment/modelo_treinado/{m}_cenario_{i}/treino_incremental_model.csv')
        df_X_Y['X_index'] = pd.to_datetime(df_X_Y['X_index'])
        df_X_Y.set_index('X_index', inplace=True)

        df_X = df_X_Y.filter(regex='^X_', axis=1)
        df_Y = df_X_Y.filter(regex='^Y_', axis=1)
        resultado_metrica = CalculoMetricas.metricas_por_dia(m, i, df_REC, df_X_Y)
        dados += resultado_metrica
        dados_tudo += resultado_metrica
    pd.DataFrame(dados).to_csv(f"../experiment/modelo_treinado/metricas_por_dia_{m}.csv", index=False)
pd.DataFrame(dados_tudo).to_csv(f"../experiment/modelo_treinado/metricas_por_dia_tudo.csv", index=False)
```
